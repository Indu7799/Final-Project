```python
import pandas as pd
from sklearn.cross_validation import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor

debug = False


def process_data(df, test=False):
    features_list = ["S1", "C1", "S2", "C2", "S3", "C3", "S4", "C4", "S5", "C5"]

    for feature in features_list:
        df = df.join(pd.get_dummies(df[feature], prefix=feature))
        df.drop(feature, 1, inplace=True)

    if not test:
        return df.drop('hand', 1), df['hand']
    else:
        return df


test = pd.read_csv('test.csv', index_col=0)
train = pd.read_csv('train.csv')

X_train, y_train = process_data(train)
X_test = process_data(test, test=True)

model = GradientBoostingRegressor(n_estimators=1500)

if debug:
    scores = cross_val_score(model, X_train, y_train, cv=5)
    print scores
    print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
else:
    model = model.fit(X_train, y_train)
    output = model.predict(X_test).astype(int)
    result = pd.Series(output, index=X_test.index)
    result.to_csv('result.csv', header=['hand'])
```
